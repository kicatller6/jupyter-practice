### 機器學習概要
機器學習其實基本上和我們一直以來說的一樣，就是我們要學一個未知的函數
 $ f(x) = y $
如果是分類，基本上就是有一筆資料 $x = (x1,x2,...,xk),$ 我們想知道這
 $f(x)=y$
其中的y就是某一個類別。
這種學函數的方法，又可以分為:
* supervised learning
* unsupervised learning
其中的supervised learning就是我們有一組知道答案的訓練資料，然後找到我們要的函數。而unsupervised learning就神了，我們不知道答案，卻要電腦自己去學!

今天我們就要來介紹最基本的方式,一個是SVM,一個是K-Means。
照例是讀入我們的標準套件。


```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

### 用SVM來分類
先做個簡單的資料
假設我們有四個點,有兩個類別。

```python
x = np.array([[-3,2],[-6,5],[3,-4],[2,-8]])
y = np.array([1,1,2,2])
```

我們要畫圖時，需要把$x$終點的$x$-座標,$y$-座標分成兩個list(array)。記得我們要$x$全部都是這樣叫出來的。

而$x$中前面的元素($x$-座標)是這樣。當然$y$-座標也是類似方法叫出來。

```python
plt.scatter([-3,-6,3,2],[2,5,-4,-8],c = y)
```

```python
x
```

```python
x[:,1]
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
```

```python
plt.scatter(x[:,0],x[:,1],c=y,s=50)
```

### SVM支持向量機
支持向量機，大家都用英文縮寫SVM稱呼，是一個用曲線把資料分隔的辦法。在高難度的時候自然就是曲面(超曲面)分隔資料的方法。

打開一台機器(這以後我們常常會做類似的動作)

```python
from sklearn.svm import SVC 
```

```python
clf = SVC()
```

```python
clf.fit(x,y)
```

```python
clf.fit(x,y)
```

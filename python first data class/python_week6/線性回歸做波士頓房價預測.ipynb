### scikit-learn真實世界數據
scikit-learn內建一些真實世界的數據,可以讓你玩玩,他們稱作'Toy Datasets'。有哪些可以參考的scikit-learn官網說明。

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
print(boston)
```

```python
boston.feature_names
```

```python
X = boston.data
Y = boston.target
```

```python
Y
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,
                                                test_size = 0.3,
                                                random_state = 87)
regr = LinearRegression()      #套用線性回歸模型
regr.fit(x_train,y_train)       #使用訓練資料
```

```python
y_predict = regr.predict(x_test)
```

```python
plt.scatter(y_test,y_predict)    #畫出測試圖形
plt.plot([0,50],[0,50],'r')
plt.xlabel('True Price')
plt.ylabel('Predicted Price')
```

```python
boston.feature_names
```

```python
plt.figure(figsize=(8,10))       
for i, feature in enumerate(boston.feature_names):      #會兩個空值i ,feature 走訪資料表
    plt.subplot(5,3,i+1)         #製表並依編號放置
    plt.scatter(X[:,i],Y,s=1)    #上點圖
    plt.ylabel('price')          #標上x,y軸項
    plt.xlabel(feature)
    plt.tight_layout()           #緊密放置製圖
```

```python
x = np.linspace(-10,10,200)
```

```python
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
```

```python
plt.subplot(221)    #subplot(ax,**kwargs) 
plt.plot(x,np.sin(x))

plt.subplot(222)
plt.plot(x,np.cos(x))

plt.subplot(223)
plt.plot(x,x)

plt.subplot(224)
plt.plot(x,x**2)
```

```python
plt.subplot(2,2,1)    #subplot(nrows,ncols,index,**kwargs) 
plt.plot(x,np.sin(x))

plt.subplot(2,2,2)
plt.plot(x,np.cos(x))

plt.subplot(2,2,3)
plt.plot(x,x)

plt.subplot(2,2,4)
plt.plot(x,x**2)
```

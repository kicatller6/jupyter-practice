# 用不同方式處理問題

```python
p = [1096.95,896.95,596.95]
```

```python
c = 30.1
```

```python
for k in p:
    print(k*c)
```

```python
result = []
```

```python
for k in p :
    twd = k*c
    result.append(twd)
print(result)
```

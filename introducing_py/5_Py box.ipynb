#### import 語法
* import ... 匯入整個檔案 用該名稱呼叫該檔案中的函式
* import ... as ... 匯入某檔案 並改變他在呼叫時使用的變數名稱
* from ... import ... 只匯入該檔案中的某函式
* from ... import ... as ... 只匯入指定檔中的指定函式並改變他在使用時的名稱
* from aaa.bbb import ... 可以利用小點來搜尋更深層的libirary內容

```python
#defaultdict 會為任何新建的鍵鍵立預設值
#int會回傳 0 list會回傳[] dict會回傳 {} 為該資料型態的預設default狀態 省略引數則會為你回傳None
from collections import defaultdict
periodic_table = defaultdict(int)   
```

```python
print(periodic_table,periodic_table[1])  #查詢任何為建立的鍵都會回傳 0 
```

```python
def no_idea():
    return 'huh?'
bestiary = defaultdict(no_idea)
bestiary['a'] = 'a'
bestiary['b'] = 'b'
print(bestiary['c'])
```

```python
# 用lambda 設定預設值
bestiary = defaultdict(lambda :'Huh?')
print(bestiary['s'])
```

```python
food_counter = defaultdict(int)    #使用初始化的int屬性 使得之後可以直接計算數字
for food in ['spam','spam','egg','spam']:
    food_counter[food] += 1
for food, count in food_counter.items():
    print(food,count)
```

```python
from collections import Counter
breakfast =['spam','spam','eggs','spam']    #
breakfast_counter = Counter(breakfast)
breakfast_counter
```

```python
breakfast_counter.most_common()  #會降冪回傳該序列
```

```python
breakfast_counter.most_common(1) #當指定數字n時會回傳n個降冪排列的序列
```

```python
lunch = ['eggs','eggs','bacon']
lunch_counter = Counter(lunch)
lunch_counter
```

```python
type(lunch_counter)
```

```python
lunch_counter + breakfast_counter
```

```python
lunch_counter - breakfast_counter #或是用其他的數學符號來做運算 例如聯集| 交集&...
```

```python
quotes = {
    'Moe':'A wise guy, huh?',
    'Larry':'Ow!',
    'Curly':'Nynk nynk!'
}
for x in quotes:
    print(x)      #在python 中的資料是無序的 (但Jupyter幫你排好了)
```

```python
from collections import OrderedDict
quotes = OrderedDict([
    ('Moe','A wise guy, huh?'),
    ('Larry','Ow!'),
    ('Curly','Nynk nynk!')
])
for x in quotes:
    print(x) 
```

```python
#deque
from collections import deque
def palidrome(word):   #回文檢查器
    
    dq = deque(word)
    while len(dq)>1:
        if dq.popleft() != dq.pop():
            return False
    return True
palidrome('sssppsss')
        
```

```python
#itertools
import itertools 
for item in itertools.chain([1,2],['a','b']): #逐一走訪目標
    print(item)
```

```python
for item in itertools.cycle([1,2]): #無限迭代器
    print(item) 
```

```python
for item in itertools.accumulate([1,2,3,4]):  #算累進值 
    print(item)
```

```python
def multiply(a,b):
    return a*b
for item in itertools.accumulate([1,2,3,4],multiply):  #也可以客製化成連續乘法
    print(item)
```
